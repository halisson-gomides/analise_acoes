### Criar Dicionários de Dataframes com cada empresa

```python
import glob
import pandas as pd

csv_files = glob.glob("datasets/originais/balancos/*.xls")
empresas = [filename.split('_')[1][:-4] for filename in csv_files]
```

```python
# fundamentos = {
#     'ABEV3': balanco_dre_abev3,
#     'MGLU3': balanco_dre_mglu3,
# }
fundamentos = {}
for i, filename in enumerate(csv_files):
    
    print(empresas[i])
    
    # pega o balanco da empresa
    balanco = pd.read_excel(filename, sheet_name=0)
    # na primeira coluna coloca o titulo com o nome da empresa
    balanco.iloc[0,0] = empresas[i]
    # tornar a 1ª linha o cabeçalho
    balanco.columns = balanco.iloc[0]
    balanco = balanco[1:]
    # tornar a 1ª coluna o index
    balanco = balanco.set_index(empresas[i])
    
    # pega o DRE da empresa e repete o processo
    dre = pd.read_excel(filename, sheet_name=1)
    # na primeira coluna coloca o titulo com o nome da empresa
    dre.iloc[0,0] = empresas[i]
    # tornar a 1ª linha o cabeçalho
    dre.columns = dre.iloc[0]
    dre = dre[1:]
    # tornar a 1ª coluna o index
    dre = dre.set_index(empresas[i])
    
    fundamentos[empresas[i]] = balanco.append(dre)

```

### Criando os rótulos: Comprar, Não Comprar ou Vender?

Não queremos saber quando vender, mas inclui essa categoria para conseguir identificar quando que o nosso modelo vai sugerir uma compra quando na verdade o melhor momento era vender. Isso significa que o modelo errou "mais" do que quando sugeriu comprar e simplesmente o certo era não comprar

Regra: 
1. Subiu mais do que o Ibovespa (ou caiu menos) -> Comprar (Valor = 2)
2. Subiu menos do que o Ibovespa até 2% (ou caiu mais do que Ibovespa até 2%) -> Não Comprar (Valor = 1)
3. Subiu menos do que o Ibovespa > 2% (ou caiu mais do que Ibovespa > 2%) -> Vender (Valor = 0)

```python
data_inicial = '12/20/2012'
data_final = '04/20/2021'

from pandas_datareader import data as web
import numpy as np

df_ibov = web.DataReader('^BVSP', data_source='yahoo', start=data_inicial, end=data_final)
df_ibov
```

```python
lst_dates = fundamentos['ABEV3'].index
for date in lst_dates:
    if date not in df_ibov.index:
        df_ibov.loc[date] = np.nan
        
df_ibov = df_ibov.sort_index()
df_ibov = df_ibov.ffill()
df_ibov = df_ibov.rename(columns={'Adj Close': 'IBOV'})

for empresa in fundamentos:
    fundamentos[empresa] = fundamentos[empresa].merge(df_ibov[['IBOV']], left_index=True, right_index=True)
    
display(fundamentos['ABEV3'])
```

###  Tornar os nossos indicadores em percentuais

```python
# fundamentos%tri = fundamentos tr / fundamentos tri anterior
# cotacao%tri = cotacao tri seguinte / cotacao tri
for empresa in fundamentos:
    fundamento = fundamentos[empresa]
    fundamento = fundamento.sort_index() # ordem crescente
    for coluna in fundamento:
        if 'Adj Close' in coluna or 'IBOV' in coluna:
            pass
        else:
            # pegar a cotação anterior -> shift(1)
            condicoes = [
                (fundamento[coluna].shift(1) > 0) & (fundamento[coluna] < 0),
                (fundamento[coluna].shift(1) < 0) & (fundamento[coluna] > 0),
                (fundamento[coluna].shift(1) < 0) & (fundamento[coluna] < 0),
                (fundamento[coluna].shift(1) == 0) & (fundamento[coluna] > 0),
                (fundamento[coluna].shift(1) == 0) & (fundamento[coluna] < 0),
                (fundamento[coluna].shift(1) < 0) & (fundamento[coluna] == 0),
            ]
            valores = [
                -1,
                1,
                (abs(fundamento[coluna].shift(1)) - abs(fundamento[coluna])) / abs(fundamento[coluna].shift(1)),
                1,
                -1,
                1
            ]
            fundamento[coluna] = np.select(condicoes, valores, default=fundamento[coluna] / fundamento[coluna].shift(1)-1)
    # pegar a cotação seguinte -> shift(-1)
    fundamento['Adj Close'] = fundamento['Adj Close'].shift(-1) / fundamento['Adj Close'] - 1
    fundamento['IBOV'] = fundamento['IBOV'].shift(-1) / fundamento['IBOV'] - 1
    fundamento['Resultado'] = fundamento['Adj Close'] - fundamento['IBOV']
    condicoes = [
        (fundamento['Resultado'] > 0),
        (fundamento['Resultado'] < 0) & (fundamento['Resultado'] >= -0.02),
        (fundamento['Resultado'] < -0.02)
    ]
    valores = [2, 1, 0]
    fundamento['Decisao'] = np.select(condicoes, valores)
    #-----------------------------------------
    
    fundamentos[empresa] = fundamento
    
display(fundamentos['ABEV3'])
```

```python
# Remover valores vazios

colunas = fundamentos['ABEV3'].columns.to_list()
valores_vazios = dict.fromkeys(colunas, 0)
total_linhas = 0

for empresa in fundamentos:

    tabela = fundamentos[empresa]
    total_linhas += len(tabela)
    qtd_null = tabela.isna().sum()
    for i, coluna in enumerate(qtd_null.index.to_list()):        
        valores_vazios[coluna] += qtd_null.values[i]

print(valores_vazios, end='\n\n')
print(f'Total de Linhas: {total_linhas}')
```

```python
# Excluir as colunas que possuem > 1/3 dos valores vazios
# Preencher com 0 o restante dos valores vazios

remover_colunas = [coluna for coluna in valores_vazios if valores_vazios[coluna] > (total_linhas / 3)]

for empresa in fundamentos:
    fundamentos[empresa] = fundamentos[empresa].drop(remover_colunas, axis=1)
    fundamentos[empresa] = fundamentos[empresa].fillna(0)
```

### Vamos partir para Feature Selection

Será que todas essas features são importantes mesmo para o nosso modelo? Muitas features nem sempre é bom, se pudermos reduzir sem perder eficiência do nosso modelo, melhor

Aqui temos 2 alternativas:

1. Seguir com todas as features e depois tentar melhorar o nosso modelo
2. Usar algum critério para selecionar as melhores features para prever e criar o modelo a partir apenas dessa seleção menor de features

Vou seguir com a opção 2, porque é mais rápida e, caso dê certo, facilita a nossa vida. Se der errado, a gente volta aqui e refaz o processo

```python
# vamos treinar um algoritmo de arvore de decisao e pegar as caracteristicas mais uteis dela
from sklearn.ensemble import ExtraTreesClassifier

modelo = ExtraTreesClassifier(random_state=1)
x = base_dados.drop('Decisao', axis=1)
y = base_dados['Decisao']
modelo.fit(x, y)

caracteristicas_importantes = pd.DataFrame(modelo.feature_importances_, x.columns).sort_values(by=0, ascending=False)
display(caracteristicas_importantes)
top10 = list(caracteristicas_importantes.index)[:10]
print(top10)
```

### Aplicação do StandardScaler para melhorar nossos modelos de MachineLearning

```python
from sklearn.preprocessing import StandardScaler

def ajustar_scaler(tabela_original):
    
    scaler = StandardScaler()
    tab_aux = tabela_original.drop('Decisao', axis=1)
#                                       valores              indices       colunas
    tab_aux = pd.DataFrame(scaler.fit_transform(tab_aux), tab_aux.index, tab_aux.columns)
    tab_aux['Decisao']  = tabela_original['Decisao']
    return tab_aux
    

base_dados_normalizada = ajustar_scaler(base_dados)
top10.append('Decisao')
base_dados_normalizada = base_dados_normalizada[top10].reset_index(drop=True)
display(base_dados_normalizada)
```

### Separação dos dados em treino e teste

```python
from sklearn.model_selection import train_test_split

x = base_dados_normalizada.drop('Decisao', axis=1)
y = base_dados_normalizada['Decisao']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, random_state=1)
```

### Criação de um Dummy Classifier (Uma baseline para ver se os nossos modelos são melhores do que puro chute)

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import classification_report, confusion_matrix

dummy = DummyClassifier(strategy='stratified', random_state=2) # chutar proporcional às categorias da base de dados
dummy.fit(x_treino, y_treino)
previsao_dummy = dummy.predict(x_teste)
```

### Métricas de Avaliação

- Precisão vai ser nossa métrica principal
- Recall pode ser útil, mas precisão no caso de ações é mt mais importante.

Explicação: Foto dos Gatos e Cachorros na Wikipedia: https://en.wikipedia.org/wiki/Precision_and_recall

```python
def avaliar(y_teste, previsoes, nome_modelo):
    
    print(nome_modelo)
    report = classification_report(y_teste, previsoes)
    print(report)
    cf_matrix = pd.DataFrame(confusion_matrix(y_teste, previsoes), index=['Vender', 'Comprar'], columns=['Vender', 'Comprar'])
    sns.heatmap(cf_matrix, annot=True, cmap='Blues', fmt=',')
    plt.show()
    print("#" * 50)
    

avaliar(y_teste, previsao_dummy, 'Dummy') # modelo de escolha aleatória

# Explicação:
# precision: - o quanto o modelo está acertando ao mandar comprar / vender
# recall: - do que era para comprar/vender, o quanto o modelo está mandando comprar/vender
# O que mais importa para esse caso é a decisão assertiva de mandar comprar: 2
```

### Modelos que vamos testar
- AdaBoost
- Decision Tree
- Random Forest
- ExtraTree
- Gradient Boost
- K Nearest Neighbors (KNN)
- Logistic Regression
- Naive Bayes
- Support Vector Machine (SVM)
- Rede Neural

```python
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

modelos = {
    'AdaBoost': AdaBoostClassifier(random_state=1),
    'DecisionTree': DecisionTreeClassifier(random_state=1),
    'RandomForest': RandomForestClassifier(random_state=1),
    'ExtraTree': ExtraTreesClassifier(random_state=1),
    'GradientBoost': GradientBoostingClassifier(random_state=1),
    'KNN': KNeighborsClassifier(),
    'LogisticRegression': LogisticRegression(random_state=1),
    'NaiveBayes': GaussianNB(),
    'SVM': SVC(random_state=1),
    'RedeNeural': MLPClassifier(random_state=1, max_iter=400),
}
```

```python
for nome_modelo in modelos:
    modelo = modelos[nome_modelo]
    modelo.fit(x_treino, y_treino)
    previsoes = modelo.predict(x_teste)
    avaliar(y_teste, previsoes, nome_modelo)
    modelos[nome_modelo] = modelo
```

**Avaliação dos modelos:**
 Os dois melhores modelos foram escolhidos por dois critérios:
- 1º - Melhor precision: RandomForest e SVM
- 2º - Melhor recall entre os dois: **RandomForest**

### Agora vamos ao tunning do modelo

- é bom sempre incluir no tuning os parâmetros "padrões" do modelo, se não pode ser que vc só encontre resultados piores

```python
from sklearn.model_selection import GridSearchCV #cross validation: validação para evitar overfitting (quando o modelo não se comporta bem para previsões de dados novos)
from sklearn.metrics import make_scorer, precision_score

# Modelo escolhido
modelo_final = modelos['RandomForest']

# 03 principais Parâmetros do RandomForest
n_estimators = range(10, 251, 30)
max_features = list(range(2, 11, 2)) # não pode ser maior que a qtd de atributos da base de dados, que no caso são 10
max_features.append('auto')
min_samples_split = range(2, 11, 2)

# score personalizado para o label 2 (comprar), que é o que nos interessa
precision2_score = make_scorer(precision_score, labels=[2], average='macro')

grid = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid={
        'n_estimators': n_estimators,
        'max_features': max_features,
        'min_samples_split': min_samples_split,
        'random_state': [1],
    },
    scoring=precision2_score, 
)

resultado_grid = grid.fit(x_treino, y_treino)
print('Ajuste feito')
```

```python
modelo_tunado = resultado_grid.best_estimator_
previsoes = modelo_tunado.predict(x_teste)
avaliar(y_teste, previsoes, 'RandomForest Tunado')
```

```python
# Verifica se essas empresas que o algoritmo recomendou a compra valeu a pena, comparando seu rendimento do ultimo trimestre
# de 2020 até o primeiro trimestre de 2021
# data_UF['NM_UF'] = data_UF['state'].map(pop_uf.set_index('UF')['NM_UF'])
tab_resultado['cotacao_inicial'] = np.nan
tab_resultado['cotacao_final'] = np.nan

for empresa in lista_empresas:
    tab_resultado.loc[tab_resultado['Empresa'] == empresa, 'cotacao_inicial'] = cotacoes[empresa].set_index('Date').loc['2020-12-31', 'Adj Close']
    tab_resultado.loc[tab_resultado['Empresa'] == empresa, 'cotacao_final'] = cotacoes[empresa].set_index('Date').loc['2021-03-31', 'Adj Close']

    
tab_resultado['rentabilidade_periodo'] = (tab_resultado['cotacao_final'] / tab_resultado['cotacao_inicial']) - 1
# Simulando um aporte de 1.000,00 no dia 31/12/2020 (fim do ultimo tri de 2020)
tab_resultado['saldo_inicial'] = 1000
tab_resultado['saldo_final'] = tab_resultado['saldo_inicial'] * (1 + tab_resultado['rentabilidade_periodo'])

display(tab_resultado)
print(f'Saldo Inicial da Carteira: {tab_resultado["saldo_inicial"].sum()}')
print(f'Saldo Final da Carteira: {tab_resultado["saldo_final"].sum()}')
print(f'Rentabilidade da Carteira: {(tab_resultado["saldo_final"].sum() / tab_resultado["saldo_inicial"].sum()) - 1}')
```
